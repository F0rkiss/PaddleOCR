```python
import cv2
import numpy as np
from ipywidgets import FileUpload
from IPython.display import display
from paddleocr import PaddleOCR
```

```python
upload = FileUpload(
    accept='image/*',
    multiple=False
)
display(upload)
```

```python
import cv2
import numpy as np

uploaded_file = upload.value[0]
content = uploaded_file['content']

img_np = np.frombuffer(content, np.uint8)
img = cv2.imdecode(img_np, cv2.IMREAD_COLOR)

print("Image loaded ✅", img.shape)
```

```python
from paddleocr import PaddleOCR

ocr_en = PaddleOCR(use_angle_cls=True, lang='en')

res_en = ocr_en.predict(img)

print("EN OCR done")
```

```python
image_path = '../data/input/sample.jpg'

# Read the image
img = cv2.imread(image_path)

# Check if image was loaded successfully
if img is None:
    print(f"Error: Could not load image from {image_path}")
    print("Please make sure the file exists and the path is correct.")
else:
    # Run OCR
    result = ocr_en.predict(img)
    
    # Check if we got any results
    if not result or not result[0]:
        print("No text detected in the image.")
    else:
        # Process and print results
        print("OCR Results:")
        print("------------")
        for idx, line in enumerate(result[0], 1):
            # Get the text and confidence
            text = line[1][0]
            conf = line[1][1] if len(line[1]) > 1 else 0.0
            
            # Get the bounding box coordinates
            bbox = line[0]
            points = np.array(bbox, dtype=np.int32)
            
            # Print the results
            print(f"Text {idx}:")
            print(f"  Content: {text}")
            print(f"  Confidence: {conf:.4f}")
            print(f"  Bounding Box: {points.tolist()}")
            print("-" * 30)
```
